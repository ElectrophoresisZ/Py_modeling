斐波那契数列：特征值和特征向量求解

```python
import sympy as sp
# 定义了一个正整数符号变量 k
sp.var('k',positive=True, integer=True)
a = sp.Matrix([[0,1],[1,1]])
val = a.eigenvals()
vec = a.eigenvects()
P,D = a.diagonalize()
ak = P@(D ** k)@(P.inv())
F = ak@sp.Matrix([1,1])
print(F[0])
#简化表达式
s = sp.simplify(F[0])
print(s)
sm = []
for i in range(20):
    # 通过 subs(k, i) 替换变量 k 为整数 i，
    # 再用 .n() 转换为数值
    sm.append(s.subs(k,i).n())
print(sm)
```
